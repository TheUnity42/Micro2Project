```python
import numpy as np
import cv2
import pytesseract
from PIL import ImageGrab
from matplotlib import pyplot as plt
```

```python
# setup tesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
```

```python
cap = ImageGrab.grab(bbox =(700, 300, 2400, 900))
img = cv2.cvtColor(np.array(cap), cv2.COLOR_BGR2GRAY)
text = pytesseract.image_to_string(img, lang='eng')
plt.imshow(img)
print(text)
```
